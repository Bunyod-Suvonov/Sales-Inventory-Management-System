## Product Details
1. Product ID
2. Name of Product
3. Quantity Available
4. Price of Product 

## Inventory.txt format
[ID],[Product Name],[Quantity Available],[Price per Quantity]

## Sales Record Details
1. Name
2. Phone Number
3. Email address
4. Product name
5. Quantity bought
6. Total price
7. Time of purchase

## Sales Record format
[Name], [Phone Number], [Email Address], [Product Name], [Quantity bought], [Time of Purchase]

```python
import time

# Reading the Inventory
fd = open('Inventory.txt','r')  
products = fd.read().split('\n')
fd.close()

# Taking User Input
ui_first_name = input("Enter your first name: ")
ui_last_name = input("Enter your last name: ")
ui_phone_number = input("Enter your phone number: ")
ui_email_address = input("Enter your email address: ")

print('We have the following products:')
print('|ID| Product name   | Quantity available | Price per unit |')
for product in products:
    product = product.split(',')
    print('|' + product[0] + ' | ' + product[1] + ' '*(15-len(product[1])) + '| ' + product[3] + ' '*(19-len(product[3])) + '| ' + product[2] + ' '*(15-len(product[2])) + '|')

ui_prod_id = input("\nEnter product ID: ")
ui_prod_qn = input("Enter product Quantity: ")

user_details = ui_first_name + ' ' + ui_last_name + ', ' + ui_phone_number + ', ' + ui_email_address





updated_product_lst = []

def printProd(prod_qn, prod_details):
    print("-----------------------------")
    print("Product Name     : ", prod_details[1])
    print("Price            : ", prod_details[2]) 
    print("Quantity         : ", prod_qn) 
    print("-----------------------------")
    print("Billing Amount   : ", int(prod_qn) * int(prod_details[2]))
    print("-----------------------------")

purchased = False
# Going through each product detail
for product in products:
    
    prod_details = product.split(',')
    
    if(prod_details[0] == ui_prod_id):
    # Checking if product exists or not

        if (int(ui_prod_qn) <= int(prod_details[3])):
        # If we're having enough quantity
        
            printProd(ui_prod_qn, prod_details)
            prod_details[3] = str(int(prod_details[3]) - int(ui_prod_qn))
            user_details = user_details + ', ' + prod_details[1] + ', ' + ui_prod_qn + ', ' + str(int(ui_prod_qn) * int(prod_details[2])) + ', ' + time.ctime()
            purchased = True
        
        elif (prod_details[3]=='0'):
        # If we don't have any at all

            print('Sorry, we do not have any quantity left for this product'+'\n'+'Please come back later')
            
        else:
        # If we're not having enough quantity
        
            print("Sorry, we do not have enough quantity.")
            print("Only",prod_details[3],'quantity available.')
            print("Would you like to purchase it?")
            
            ch = input("Press Y/N: ")
            
            if (ch == 'Y' or ch == 'y'):
            # If you want to purchase with remaining quantity
                printProd(prod_details[3], prod_details)
                user_details = user_details + ', ' + prod_details[1] + ', ' + prod_details[3] + ', ' + str(int(prod_details[3]) * int(prod_details[2])) + ', ' + time.ctime()
                prod_details[3] = '0'
                purchased = True
            
    # Updating my Inventory List
    updated_product_lst.append(prod_details)


# Store user details
if purchased==True:
    fd = open('sales_records.txt', 'r')
    users = fd.read()
    fd.close()
    fd = open('sales_records.txt', 'w')
    if len(users) > 1:
        user_details = '\n' + user_details
        users = users + user_details
    else:
        users = users + user_details
    fd.write(users)
    fd.close()

    # Updating my Inventory String
    lst = []
    
    for i in updated_product_lst:
        prod = i[0] +","+  i[1] +","+ i[2] +","+ i[3] + '\n'
        lst.append(prod)    
    lst[-1] = lst[-1][:-1]
    

# Updating Inventory File
fd = open('Inventory.txt','w')

for i in lst:
    fd.write(i)

fd.close()

print('Thanks for your shopping! Hope to see you back!')
```
