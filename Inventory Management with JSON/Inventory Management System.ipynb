```python
import time
import json

ch = 'Y'
fd = open('Products.json','r')
js = fd.read()
fd.close()

record = json.loads(js)

print("--------------------MENU---------------------")
for key in record.keys():
    print(key, record[key]['name'], record[key]['price'], record[key]['quantity'])
print("---------------------------------------------")
print('')


ui_name = str(input("Enter your name    : "))
ui_mail = str(input("Enter Mail ID      : "))
ui_ph   = str(input("Enter Phone No     : "))
ui_pr   = str(input("Enter product ID   : "))
ui_qn   = int(input("Enter Quantity     : "))

print("---------------------------------------------")
print('')

if (record[ui_pr]['quantity'] >= ui_qn):

    print("Name      : ", record[ui_pr]["name"])
    print("Price ($) : ", record[ui_pr]["price"])
    print("Quantity  : ", ui_qn)
    print("---------------------------------------------")
    print("Billing   : ", ui_qn * record[ui_pr]["price"], "$")
    print("---------------------------------------------")

    record[ui_pr]['quantity'] = record[ui_pr]['quantity'] - ui_qn
    
    sale = ui_name+","+ui_mail+","+ui_ph+","+ui_pr+","+record[ui_pr]["name"]+","+str(ui_qn)+","+str(record[ui_pr]["price"])+","+str(ui_qn * record[ui_pr]["price"])+","+time.ctime()+"\n"

else:
    
    print("Sorry, We're not having enough quanity of product in our Inventory.")
    print("We're only having " + str(record[ui_pr]['quantity']) + " quantity.")
    print("---------------------------------------------")
    
    ch == str(input("Press Y to purchase: "))
    
    if(ch == "Y" or ch == 'y'):

        print("---------------------------------------------")
        print("Name      : ", record[ui_pr]["name"])
        print("Price (Rs): ", record[ui_pr]["price"])
        print("Quantity  : ", record[ui_pr]['quantity'])
        print("---------------------------------------------")
        print("Billing   : ", record[ui_pr]['quantity'] * record[ui_pr]["price"], "$")
        print("---------------------------------------------")

        record[ui_pr]['quantity'] = 0
        
        sale = ui_name+","+ui_mail+","+ui_ph+","+ui_pr+","+record[ui_pr]["name"]+","+str(record[ui_pr]['quantity'])+","+str(record[ui_pr]["price"])+","+str(record[ui_pr]['quantity'] * record[ui_pr]["price"])+","+time.ctime()+"\n"
        
    else:
        print("Thanks!")
        
js = json.dumps(record)

fd = open('Products.json','w')
fd.write(js)
fd.close()

fd = open('Sales.txt','a')
fd.write(sale)
fd.close()

print('')
print("---------------------------------------------")
print("  Thanks for your order, Inventory Updated!  ")
print("---------------------------------------------")











```
